```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
train_df = pd.read_csv("/kaggle/input/summer-analytics-mid-hackathon/hacktrain.csv")
test_df = pd.read_csv("/kaggle/input/summer-analytics-mid-hackathon/hacktest.csv")

train_df
```

```python
# let's use limited columns which makes more sense for serving our purpose

cols_to_use = ['Unnamed: 0', 'ID', 'class', '20150720_N', '20150602_N', '20150517_N',
       '20150501_N', '20150415_N', '20150330_N', '20150314_N', '20150226_N',
       '20150210_N', '20150125_N', '20150109_N', '20141117_N', '20141101_N',
       '20141016_N', '20140930_N', '20140813_N', '20140626_N', '20140610_N',
       '20140525_N', '20140509_N', '20140423_N', '20140407_N', '20140322_N',
       '20140218_N', '20140202_N', '20140117_N', '20140101_N']

dataset = train_df[cols_to_use]
dataset
```

```python
dataset.isna().sum()
```

```python

ndvi_cols = [col for col in dataset.columns if '_N' in col]
dataset[ndvi_cols] = dataset[ndvi_cols].interpolate(axis=1).bfill(axis=1).ffill(axis=1)
dataset['ndvi_mean'] = dataset[ndvi_cols].mean(axis=1)
dataset['ndvi_std'] = dataset[ndvi_cols].std(axis=1)
dataset['ndvi_range'] = dataset[ndvi_cols].max(axis=1) - dataset[ndvi_cols].min(axis=1)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
# y = le.fit_transform(df['class'])
dataset['class'] = le.fit_transform(dataset['class'])
```

```python
dataset.isna().sum()
```

```python
X = dataset.iloc[:, 0:1]
y = dataset.iloc[:, 1]
X = dataset.drop(['ID', 'class'], axis=1)
y = dataset['class']
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, stratify=y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000, multi_class='multinomial', solver='lbfgs')
model.fit(train_X, train_y)
```

```python
train_X.shape
```

```python
train_y.shape
```

```python
from sklearn.metrics import accuracy_score, classification_report
y_pred = model.predict(test_X) #calculating prediction score 
#by first predicting and then calculating also called accuracy
print("Accuracy:", accuracy_score(test_y, y_pred))
print(classification_report(test_y, y_pred))
```

```python
# Make Final Predictions
test_df = pd.read_csv('/kaggle/input/summer-analytics-mid-hackathon/hacktest.csv')



test_df[ndvi_cols] = test_df[ndvi_cols].interpolate(axis=1).bfill(axis=1).ffill(axis=1)
test_df['ndvi_mean'] = test_df[ndvi_cols].mean(axis=1)
test_df['ndvi_std'] = test_df[ndvi_cols].std(axis=1)
test_df['ndvi_range'] = test_df[ndvi_cols].max(axis=1) - test_df[ndvi_cols].min(axis=1)



preds = model.predict(test_df.drop(['ID'], axis=1))
test_df['class'] = le.inverse_transform(preds)
test_df[['ID', 'class']].to_csv('submission.csv', index=False)
```
